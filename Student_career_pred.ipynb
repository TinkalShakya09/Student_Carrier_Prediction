```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("Student.csv")
```

```python
x= data[["Hindi","English","Mathematics","Biology","Chemistry","pyhsics","Computer Science","Art","Accountancy","Physical Education","General Knowledge","History"]]
```

```python
y=data[["Doctor","Engineering","CA/B.Com","UPSC/BA"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from keras.models import Sequential
```

```python
model=Sequential()
```

```python
from keras.layers import Dense
```

```python
model.add(Dense(units=5, input_shape=(12, ), activation='relu', kernel_initializer='he_normal' ))
model.add(Dense(units=8,  activation='relu', kernel_initializer='he_normal' ))
model.add(Dense(units=4,  activation='softmax'))#, kernel_initializer='zeros', ))
```

```python
from keras.optimizers import RMSprop
```

```python
model.compile(optimizer=RMSprop(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=300,shuffle=True,verbose=2)
```

```python
y_pred=model.predict([[0, 1, 1, 0, 0, 0, 1, 0, 0, 0, 0, 0]])
y_pred
```

```python
model.save("model.h5")
print("Saved model to disk")
```
